#### incident_ward is the only attribute with missing values.
- We will use the median value to fill in missing values

```python
fire_num_df = fire_df[['num_casualties', 'num_rescued', 'est_loss', 'est_displaced', 
                       'incident_station_area', 'incident_ward', 'latitude', 'longitude', 
                       'response_time']]
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
```

```python
imputer.fit(fire_num_df)
```

```python
imputer.statistics_
```

```python
fire_num_df.median().values
```

```python
X = imputer.transform(fire_num_df)
```

```python
fire_num_df = pd.DataFrame(X, columns=fire_num_df.columns, index=fire_num_df.index)
```
